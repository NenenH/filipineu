```python
import subprocess
```

```python
source = open('source.data.eval', 'r')
# output = open('smt-result.data.eval', 'w')
output = []

data = source.readlines()

for datum in data:
    command = "thot_client -i localhost -t '" + datum + "'"
    p = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.STDOUT)
    for line in p.stdout.readlines():
        print(line.decode('utf-8'))
#         output.write(line.decode('utf-8') + "\n")
        output.append(line.decode('utf-8'))
    retval = p.wait()

source.close()
```

```python
file = open('smt-result.data.eval', 'w')

for out in output:
    if out != "\n":
        file.write(out)
    
file.close()
```
